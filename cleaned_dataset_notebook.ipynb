```python
import pandas as pd
import numpy as np
```

```python
file_path = "Rheumatic and Autoimmune Disease Dataset (1).xlsx"
data = pd.read_excel(file_path)


print("Initial Dataset Shape:", data.shape)
data.head()
```

```python
filtered_data = data[data['Disease'].isin(['Rheumatoid Arthritis', 'Systemic Lupus Erythematosus'])].copy()
print("Filtered Dataset Shape:", filtered_data.shape)
filtered_data['Disease'].value_counts()
```

### 🧬 Step 3: Handling Missing Values in Categorical Columns
Medical test columns like ANA, Anti-Ro, etc. can be **Positive** or **Negative**.  
If a value is missing, it usually means the test wasn’t done or returned normal — so we fill them as `"Negative"`.


```python
categorical_cols = ['HLA-B27', 'ANA', 'Anti-Ro', 'Anti-La', 'Anti-dsDNA', 'Anti-Sm']
filtered_data[categorical_cols] = filtered_data[categorical_cols].fillna('Negative')
```

### 🧪 Step 4: Handling Missing Values in Numerical Columns
We fill missing lab values (like ESR, CRP, RF, Anti-CCP, C3, C4) with the **median per disease group**.

This approach:
- Preserves disease-level variation.
- Avoids distortion caused by extreme values.
- Keeps the data medically realistic.


```python
numerical_cols = ['ESR', 'CRP', 'RF', 'Anti-CCP', 'C3', 'C4']
for col in numerical_cols:
    filtered_data[col] = filtered_data.groupby('Disease')[col].transform(lambda x: x.fillna(x.median()))
```

```python
filtered_data.isnull().sum()
```

```python
output_path = "Cleaned_Rheumatic_Lupus_Dataset.xlsx"
filtered_data.to_excel(output_path, index=False)
print("✅ Cleaned dataset saved successfully as:", output_path)
```
